```python
# =============================================================================
# SECCIÃ“N 2: CARGA DE DATOS Y ARTEFACTOS EVOLUCIONADOS
# =============================================================================
import torch
import os
import requests
import numpy as np

# ConfiguraciÃ³n de URLs y Directorios
BASE_URL_DATA = "https://raw.githubusercontent.com/berniehans/ProjectEvo-Callao-Safety/main/data_processed/"
BASE_URL_MODELS = "https://raw.githubusercontent.com/berniehans/ProjectEvo-Callao-Safety/main/models/"
DIRS = ["data_processed", "models"]

for d in DIRS: os.makedirs(d, exist_ok=True)

def download(url, path):
    if not os.path.exists(path):
        print(f"ðŸ“¥ Descargando {path}...", end=" ")
        try:
            r = requests.get(url)
            if r.status_code == 200:
                with open(path, "wb") as f: f.write(r.content)
                print("âœ…")
            else: print(f"âŒ Error {r.status_code}")
        except Exception as e: print(f"âš ï¸ {e}")

# 1. Descargar Datos Fragmentados (V2)
download(BASE_URL_DATA + "metadata.pt", "data_processed/metadata.pt")
meta = torch.load("data_processed/metadata.pt", weights_only=False)

for i in range(1, meta['total_chunks'] + 1):
    download(BASE_URL_DATA + f"bert_embeddings_part_{i}.pt", f"data_processed/bert_embeddings_part_{i}.pt")

# 2. Descargar MÃ¡scara (V3) y ConfiguraciÃ³n NAS (V4)
download(BASE_URL_MODELS + "ga_feature_mask.pt", "models/ga_feature_mask.pt")
download(BASE_URL_MODELS + "best_nas_config.pt", "models/best_nas_config.pt")

# 3. ReconstrucciÃ³n del Dataset
X_list, y_list = [], []
for i in range(1, meta['total_chunks'] + 1):
    d = torch.load(f"data_processed/bert_embeddings_part_{i}.pt", weights_only=False)
    X_list.append(d['embeddings']); y_list.append(d['labels'])

X_all = torch.cat(X_list, dim=0)
y_all = torch.cat(y_list, dim=0)

# 4. Aplicar la MÃ¡scara Evolutiva (V3)
mask_data = torch.load("models/ga_feature_mask.pt", weights_only=False)
mask = torch.tensor(mask_data['best_mask']).bool()
X_final = X_all[:, mask]

# 5. Cargar la Arquitectura Ganadora (V4)
nas_config = torch.load("models/best_nas_config.pt", weights_only=False)
best_arch = nas_config['best_solution'] # [layers, neurons, dropout, lr, alpha]

# Preparar GPU
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
X_gpu = X_final.to(device).half()
y_gpu = y_all.to(device).long()

print("="*40)
print(f"ðŸš€ SISTEMA LISTO PARA ENSEMBLE")
print(f"ðŸ”¹ Datos: {X_gpu.shape}")
print(f"ðŸ”¹ Arquitectura Base: {int(best_arch[0])} capas x {int(best_arch[1])} neuronas")
print("="*40)
```

```python
# =============================================================================
# SECCIÃ“N 3: ARQUITECTURA DEL CLASIFICADOR (BASE LEARNER)
# =============================================================================
import torch.nn as nn

class EvoClassifier(nn.Module):
    def __init__(self, input_dim, dropout, output_dim):
        super(EvoClassifier, self).__init__()
        # Arquitectura Fija: 3 Capas Ocultas de 512 Neuronas
        # El Ãºnico hiperparÃ¡metro variable aquÃ­ es el Dropout
        self.net = nn.Sequential(
            nn.Linear(input_dim, 512), nn.ReLU(), nn.Dropout(dropout),
            nn.Linear(512, 512), nn.ReLU(), nn.Dropout(dropout),
            nn.Linear(512, 512), nn.ReLU(), nn.Dropout(dropout),
            nn.Linear(512, output_dim)
        )

    def forward(self, x):
        return self.net(x)
```

```python
# =============================================================================
# SECCIÃ“N 4: FITNESS FUNCTION - TURBO GPU (CORREGIDO)
# =============================================================================
from sklearn.metrics import f1_score
from sklearn.model_selection import StratifiedKFold
import torch
import torch.nn as nn
import numpy as np

# 1. PREPARACIÃ“N ESTÃTICA (Residente en GPU)
print("ðŸš€ Cargando datos y tensores estÃ¡ticos a VRAM...", end=" ")

# Datos completos en GPU (Float32 para estabilidad)
X_gpu_full = X_final.to(device).float()
y_gpu_full = y_all.to(device).long()

# Pre-calcular Folds (Indices en CPU por ahora)
skf = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)
# Convertimos y_all a numpy explÃ­citamente por si acaso
FOLDS_INDICES = list(skf.split(X_final.cpu().numpy(), y_all.cpu().numpy()))

print("âœ… Listo.")

# ConfiguraciÃ³n de Velocidad
BATCH_SIZE = 8192  # Mega-Lote
EPOCHS_FAST = 15   # Ã‰pocas de tuning
CRITICAL_INDICES = [3, 4, 5]

def fitness_turbo(ga_instance, solution, solution_idx):
    dropout, lr, alpha = solution
    print(".", end="", flush=True) # Latido visual

    scores = []

    try:
        for train_idx_np, val_idx_np in FOLDS_INDICES:

            # --- CORRECCIÃ“N CLAVE: INDICES A GPU ---
            # Movemos los Ã­ndices del fold a la GPU una sola vez
            train_idx = torch.tensor(train_idx_np, device=device)
            # val_idx no lo necesitamos en GPU para slicing masivo, usamos numpy abajo

            # Instanciar Modelo
            model = EvoClassifier(X_gpu_full.shape[1], float(dropout), len(meta['class_names'])).to(device)

            # Pesos y Optimizador
            w = torch.ones(len(meta['class_names']), device=device)
            for c in CRITICAL_INDICES: w[c] = float(alpha)

            crit = nn.CrossEntropyLoss(weight=w)
            opt = torch.optim.Adam(model.parameters(), lr=float(lr))

            model.train()
            n_samples = len(train_idx)

            # Generamos permutaciÃ³n aleatoria EN GPU
            shuffled_idx = torch.randperm(n_samples, device=device)

            for _ in range(EPOCHS_FAST):
                for i in range(0, n_samples, BATCH_SIZE):
                    # Todo esto ocurre dentro de la GPU ahora:
                    # 1. Obtenemos Ã­ndices del batch (GPU)
                    batch_idx_local = shuffled_idx[i : i + BATCH_SIZE]

                    # 2. Obtenemos Ã­ndices reales del dataset (GPU a GPU)
                    batch_real_idx = train_idx[batch_idx_local]

                    # 3. Slicing de datos (GPU a GPU)
                    bx = X_gpu_full[batch_real_idx]
                    by = y_gpu_full[batch_real_idx]

                    opt.zero_grad()
                    crit(model(bx), by).backward()
                    opt.step()

            # --- EVALUACIÃ“N ---
            model.eval()
            with torch.no_grad():
                # Inferencia sobre validaciÃ³n (usando Ã­ndices numpy para slice directo si se prefiere)
                # O pasamos val_idx a tensor para mantener consistencia. Usemos el mÃ©todo directo:
                val_out = model(X_gpu_full[val_idx_np])
                val_preds = val_out.argmax(1).cpu().numpy()

                # Ground truth desde CPU (seguro)
                val_true = y_all[val_idx_np].cpu().numpy()

            scores.append(f1_score(val_true, val_preds, average='macro', zero_division=0))

        return np.mean(scores)

    except Exception as e:
        print(f"X({e})", end="")
        return 0.0
```

```python
# =============================================================================
# SECCIÃ“N 5: EJECUCIÃ“N OPTIMIZADA
# =============================================================================
import pygad
from tqdm.auto import tqdm

gene_space = [
    [0.2, 0.3, 0.4, 0.5],       # Dropout
    [0.0005, 0.001, 0.002],     # Learning Rate
    [3.0, 5.0, 8.0, 10.0, 15.0] # Alpha
]

n_gen = 25
pbar = tqdm(total=n_gen, desc="ðŸš€ Tuning V5 (GPU Turbo)")

def on_gen(ga):
    pbar.update(1)
    best = np.max(ga.last_generation_fitness)
    print(f" -> Gen OK. Best F1: {best:.4f}")
    pbar.set_postfix({"Best F1": f"{best:.4f}"})

ga_tune = pygad.GA(
    num_generations=n_gen,
    sol_per_pop=12,
    num_parents_mating=5,
    fitness_func=fitness_turbo,
    num_genes=3,
    gene_space=gene_space,
    parent_selection_type="tournament",
    mutation_type="random",
    mutation_percent_genes=40,
    on_generation=on_gen,
    keep_elitism=2
)

print("ðŸ”¥ Iniciando Tuning a Alta Velocidad...")
ga_tune.run()
pbar.close()

# Resultados
best_sol, best_fit, _ = ga_tune.best_solution()
print(f"\nðŸ† MEJOR CONFIGURACIÃ“N (V5):")
print(f"   Dropout: {best_sol[0]}")
print(f"   Learning Rate: {best_sol[1]}")
print(f"   Alpha: {best_sol[2]}")
print(f"   F1-Macro (CV): {best_fit:.4f}")
```

```python
# =============================================================================
# SECCIÃ“N 6: ENTRENAMIENTO FINAL (ENSEMBLE) Y EVALUACIÃ“N EN TEST
# =============================================================================
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

print("ðŸ”§ Preparando Fase Final...")

# 1. GENERACIÃ“N ROBUSTA DE DATASETS
# Recuperamos los Ã­ndices originales. Usamos X_final.
indices = np.arange(len(X_final))
idx_train_full, idx_test = train_test_split(indices, test_size=0.15, stratify=y_all, random_state=42)

print(f"âœ… Datos Listos: {len(idx_train_full)} para Entrenamiento / {len(idx_test)} para Test Final")

# 2. ConfiguraciÃ³n Ganadora (Recuperada de SecciÃ³n 5)
try:
    DROPOUT_OPT, LR_OPT, ALPHA_OPT = best_sol
except NameError:
    print("âš ï¸ No se encontraron genes del GA, usando valores por defecto.")
    DROPOUT_OPT, LR_OPT, ALPHA_OPT = 0.4, 0.002, 8.0

# 3. Entrenamiento del ComitÃ© de Expertos (5 Modelos)
models_pool = []
# Preparamos datos de entrenamiento en GPU (Float32)
X_train_gpu = X_final[idx_train_full].to(device).float()
y_train_gpu = y_all[idx_train_full].to(device).long()

loader = DataLoader(
    TensorDataset(X_train_gpu, y_train_gpu),
    batch_size=2048, shuffle=True
)

# Pesos de clase Ã³ptimos
class_weights = torch.ones(len(meta['class_names']), device=device)
for c in [3, 4, 5]: class_weights[c] = float(ALPHA_OPT)

print(f"\nðŸ”¥ Entrenando Ensemble (5 Expertos) con Alpha={ALPHA_OPT}...")

for i in range(5):
    torch.manual_seed(i * 100) # Semillas distintas para variabilidad (Bagging)

    # Instanciar Modelo con Dropout Ã“ptimo
    model = EvoClassifier(X_final.shape[1], float(DROPOUT_OPT), len(meta['class_names'])).to(device)
    opt = torch.optim.Adam(model.parameters(), lr=float(LR_OPT))
    crit = nn.CrossEntropyLoss(weight=class_weights)

    model.train()
    # Entrenamos por 30 Ã©pocas para asegurar convergencia total
    for ep in range(30):
        for bx, by in loader:
            opt.zero_grad(); crit(model(bx), by).backward(); opt.step()

    model.eval()
    models_pool.append(model)
    print(f"   âœ… Experto #{i+1} entrenado.")

# 4. Inferencia en Test Set (Ensemble Voting)
print("\nðŸ” Evaluando en Test Set (Datos nunca vistos)...")
# Preparamos test en GPU
X_test_gpu = X_final[idx_test].to(device).float()

test_probs = []
with torch.no_grad():
    for model in models_pool:
        # Probabilidades Softmax de cada experto
        probs = torch.softmax(model(X_test_gpu), dim=1).cpu().numpy()
        test_probs.append(probs)

# Promedio de las predicciones (Soft Voting)
avg_probs = np.mean(test_probs, axis=0)
y_pred_final = np.argmax(avg_probs, axis=1)
y_true_final = y_all[idx_test].numpy()

# 5. Reporte de Resultados
print("\n" + "="*60)
print("ðŸ“Š REPORTE OFICIAL DE TESIS (V5 POTENCIADA)")
print("="*60)
print(classification_report(y_true_final, y_pred_final, target_names=meta['class_names']))

# Matriz de ConfusiÃ³n
plt.figure(figsize=(12, 8))
cm = confusion_matrix(y_true_final, y_pred_final, normalize='true')
sns.heatmap(cm, annot=True, fmt=".2f", cmap="Greens",
            xticklabels=meta['class_names'], yticklabels=meta['class_names'])
plt.title(f"Matriz de ConfusiÃ³n (Ensemble 3x512 | Alpha={ALPHA_OPT})")
plt.xlabel("PredicciÃ³n")
plt.ylabel("Realidad")
plt.show()

# Guardado Final
torch.save({
    'models': [m.state_dict() for m in models_pool],
    'config': [DROPOUT_OPT, LR_OPT, ALPHA_OPT],
    'arch': "3x512"
}, "models/production_final.pt")
print("ðŸ’¾ Modelo final guardado en 'models/production_final.pt'")
```
